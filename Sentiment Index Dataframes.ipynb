```python
import pandas as pd 
import numpy as np
import pickle 
```

```python
# load pickles 
tweets_pkl = open("tweets.pkl", "rb")
tweets_df = pickle.load(tweets_pkl)

tweets_sent_pkl = open("tweets_sent.pkl", "rb")
tweets_sent = pickle.load(tweets_sent_pkl)

stocks_pkl = open("stocks.pkl", "rb")
stocks = pickle.load(stocks_pkl)

tweet_sent_pkl = open("stock_tweet_sentiment.pkl", "rb")
tweets = pickle.load(tweet_sent_pkl)

reviews_pkl = open("reviews.pkl", "rb")
ratings = pickle.load(reviews_pkl)

news_sent_pkl = open("news_sent.pkl", "rb")
news = pickle.load(news_sent_pkl)
```

### Dataframe with yearly averages of all categories

#### Reformatting 

```python
# reformat tweets
tweets = tweets_sent.copy()
tweets["Date"] = tweets["Date"].dt.date
tweets["Date"] = pd.to_datetime(tweets["Date"])
tweets = tweets[["Tweet","Date","Sentiment"]]
tweets.rename(columns={"Sentiment":"Tweet_Sentiment"},inplace=True)
```

```python
tweets.head()
```

```python
news.head()
```

```python
# reformat news
news["Date"] = pd.to_datetime(news["Date"]) 
news.sort_values("Date",inplace=True)
```

```python
ratings.head()
```

```python
# reformat ratings
ratings["rating"] = ratings["rating"].astype(str)
scaled_sent = {"1.0":"-1", "2.0":"-.5", "3.0":"0", "4.0":".5", "5.0":"1"}
ratings.replace({"rating":scaled_sent},inplace=True)
ratings["rating"] = (ratings["rating"]).astype(float)
ratings.rename(columns={"rating":"Employee_Sentiment"},inplace=True)
ratings.rename(columns={"date":"Date"},inplace=True)
ratings.rename(columns={"date":"Date", "rating":"Rating"},inplace=True)
ratings.head()
```

```python
ratings.sort_values("Date",inplace=True)
tweets.sort_values("Date",inplace=True)
```

```python
stocks.head()
```

```python
# reformat stocks 
stocks["Value"] = (stocks["Open"]+stocks["High"]+stocks["Low"]+stocks["Adj Close"]/4)
stocks.drop(columns=["Open", "High", "Low", "Volume", "Adj Close"],inplace=True)
stocks.head()
```

```python
# combine the data and figure out the nans 
sentidx_df = pd.concat([stocks,tweets,ratings,news])
sentidx_df
```

```python
# sort by dates
# sentidx_df["Year"] = sentidx_df["Date"].dt.year
sentidx_df["Month"] = sentidx_df["Date"].dt.month
```

```python
sentidx_df = (sentidx_df.groupby(["Year","Month"])[["Employee_Sentiment", "News_Sentiment", "Tweet_Sentiment", "Value"]].mean()).reset_index()
```

```python
sentidx_df = sentidx_df[sentidx_df["Year"] >= 2012]
```

```python
sentidx_df = sentidx_df[sentidx_df["Year"] < 2018]
```

```python
sentidx_df.fillna(method="bfill",inplace=True)
```

```python
sentidx_df.fillna(method="ffill",inplace=True)
```

```python
sentidx_df
```

```python
# make sentiment index
sentidx_df["Sentiment_Index"] = (sentidx_df["Employee_Sentiment"] + sentidx_df["News_Sentiment"] + sentidx_df["Tweet_Sentiment"]) / 3

```
